```python
"""SQL solution"""
SELECT dept_no, GROUP_CONCAT(emp_no SEPARATOR ',')
FROM dept_emp
GROUP BY dept_no
```

```python
"""pandas dataframe creation"""
import pandas as pd
# read records from a txt file
dept_emp = pd.read_csv('data.txt', header=None)
dept_emp.columns = ['emp_no', 'dept_no', 'from_date', 'to_date']
dept_emp.head()
```

```python
# clean the records
dept_emp['emp_no'] = dept_emp['emp_no'].str[-5:]
for col in ['dept_no', 'from_date']:
    dept_emp[col] = dept_emp[col].str[1:-1]
dept_emp['to_date'] = dept_emp['to_date'].str[1:-3]
dept_emp.head()
```

```python
df = dept_emp[['emp_no', 'dept_no']]
df['emps'] = df.groupby('dept_no')['emp_no'].transform(lambda x: ','.join(x))
df = df[['dept_no', 'emps']].drop_duplicates()
df
```
