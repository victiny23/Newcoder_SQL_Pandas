```python
"""SQL solution"""
# 1
SELECT *
FROM employees
ORDER BY hire_date DESC
LIMIT 1
OFFSET 2

# 2
SELECT r.emp_no, r.birth_date, r.first_name, r.last_name, r.gender, r.hire_date
FROM
(SELECT *, ROW_NUMBER() OVER(ORDER BY hire_date DESC) AS rk
FROM employees) AS r
WHERE r.rk = 3
```
