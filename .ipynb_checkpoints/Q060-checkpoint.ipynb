```python
"""SQL solution"""
SELECT emp_no, salary, SUM(salary) OVER(ORDER BY emp_no) AS running_total
FROM salaries
WHERE to_date = '9999-01-01'
```

```python
"""pandas dataframe creation"""
import pandas as pd

# read data from a .txt file
salaries = pd.read_csv('data.txt', header=None)
salaries.columns = ['emp_no', 'salary', 'from_date', 'to_date']
salaries.head()
```

```python
# clean the data
salaries['emp_no'] = salaries['emp_no'].str[-5:]
salaries['from_date'] = salaries['from_date'].str[1:-1]
salaries['to_date'] = salaries['to_date'].str[1:-3]
salaries.head()
```

```python
df = salaries[salaries['to_date'] == '9999-01-01']
df = df[['emp_no', 'salary']]
df['running_total'] = df['salary'].cumsum()
df
```
