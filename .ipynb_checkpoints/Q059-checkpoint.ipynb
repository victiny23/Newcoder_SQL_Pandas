```python
"""SQL solution"""
# 1
SELECT e.emp_no, first_name, last_name, b.btype, salary,
(CASE
    WHEN b.btype = 1 THEN salary * 0.1
    WHEN b.btype = 2 THEN salary * 0.2
    ELSE salary * 0.3
END)
FROM employees AS e
INNER JOIN emp_bonus AS b
ON e.emp_no = b.emp_no
INNER JOIN salaries AS s
ON e.emp_no = s.emp_no
WHERE s.to_date = '9999-01-01'

# 2
SELECT e.emp_no, first_name, last_name, b.btype, salary,
CASE
    WHEN b.btype = 1 THEN salary * 0.1
    WHEN b.btype = 2 THEN salary * 0.2
    ELSE salary * 0.3
END
FROM employees AS e, emp_bonus AS b, salaries AS s
WHERE e.emp_no = b.emp_no
AND e.emp_no = s.emp_no
AND s.to_date = '9999-01-01'
```

```python
"""pandas dataframe creation"""
import pandas as pd

# read records from a .txt file
emp_bonus = pd.read_csv('data.txt', header=None)
emp_bonus.columns = ['emp_no', 'received', 'btype','']
emp_bonus = emp_bonus.iloc[:,:3]
emp_bonus.head()
```

```python
# clean the data
emp_bonus['emp_no'] = emp_bonus['emp_no'].str[1:]
emp_bonus['received'] = emp_bonus['received'].str[2:-1]
emp_bonus['btype'] = emp_bonus['btype'].str[0]
emp_bonus.head()
```

```python
# read records from a .txt file
employees = pd.read_csv('data.txt', header=None)
employees.columns = ['emp_no', 'birth_date', 'first_name', 'last_name', 'gender', 'hire_date']
employees.head()
```

```python
# clean the records
employees['emp_no'] = employees['emp_no'].str[-5:]
for col in ['birth_date', 'first_name', 'last_name', 'gender']:
    employees[col] = employees[col].str[1:-1]
employees['hire_date'] = employees['hire_date'].str[1:-3]
employees.head()
```

```python
# read records from a .txt file
salaries = pd.read_csv('data.txt', header=None)
salaries.columns = ['emp_no', 'salary', 'from_date', 'to_date']
salaries.head()
```

```python
# clean the data
salaries['emp_no'] = salaries['emp_no'].str[-5:]
salaries['from_date'] = salaries['from_date'].str[1:-1]
salaries['to_date'] = salaries['to_date'].str[1:-3]
salaries.head()
```

```python
print(len(salaries))
df = salaries[salaries['to_date'] == '9999-01-01']
print(len(df))
df.head()
```

```python
df = pd.merge(df[['emp_no', 'salary']], employees[['emp_no', 'first_name', 'last_name']], on='emp_no', how='left')
df.head()
```

```python
df = pd.merge(df, emp_bonus[['emp_no', 'btype']], on='emp_no', how='inner')
df.head()
```

```python
df['btype'] = df['btype'].astype('int')
df['bonus'] = df['salary'] * 0.3
df.loc[df['btype']==1, 'bonus'] = df.loc[df['btype']==1, 'salary'] * 0.1
df.loc[df['btype']==2, 'bonus'] = df.loc[df['btype']==2, 'salary'] * 0.2
df[['emp_no', 'first_name', 'last_name', 'btype', 'salary', 'bonus']]
```
