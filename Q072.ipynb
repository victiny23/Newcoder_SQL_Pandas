```python
"""Find the average score for each job, order by score descending with 3 decimals"""
```

```python
"""SQL table creation and data insertion"""
drop table if exists grade;

CREATE TABLE  grade(
`id` int(4) NOT NULL,
`job` varchar(32) NOT NULL,
`score` int(10) NOT NULL,
PRIMARY KEY (`id`));

INSERT INTO grade VALUES
(1,'C++',11001),
(2,'C++',10000),
(3,'C++',9000),
(4,'Java',12000),
(5,'Java',13000),
(6,'JS',12000),
(7,'JS',11000),
(8,'JS',9999);
```

```python
"""SQL solution"""
SELECT job, ROUND(AVG(score), 3) AS avgs
FROM grade
GROUP by job
ORDER BY avgs DESC
```

```python
"""pandas dataframe creation"""
import pandas as pd

grade = pd.read_csv('data.txt', header=None)
grade
```

```python
# clean the dataframe
grade = grade.drop(columns=3)
grade.columns = ['id', 'job', 'score']
grade['id'] = grade['id'].str[1:]
grade['job'] = grade['job'].str[1:-1]
grade['score'] = grade['score'].str[:-1]
grade.iloc[-1,2] = grade.iloc[-1,2][:-1]
grade
```

```python
grade['score'] = grade['score'].astype('int')
grade.groupby('job')['score'].mean().reset_index().sort_values('score', ascending=False)
```
