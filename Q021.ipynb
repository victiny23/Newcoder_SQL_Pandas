```python
"""SQL solution"""
# 1
SELECT c.emp_no, c.cur - i.ini AS growth
FROM 
(SELECT emp_no, salary AS cur
FROM salaries
WHERE to_date = '9999-01-01') AS c
INNER JOIN
(SELECT e.emp_no, s.salary AS ini
FROM employees AS e
INNER JOIN salaries AS s
ON e.emp_no = s.emp_no
AND e.hire_date = s.from_date) AS i
ON c.emp_no = i.emp_no
ORDER BY growth

# 2
SELECT s_now.emp_no, s_now.salary - s_hire.salary
FROM
(SELECT s1.emp_no, s1.salary
FROM employees AS e1
INNER JOIN salaries AS s1
ON e1.emp_no = s1.emp_no 
AND e1.hire_date = s1.from_date
WHERE s1.emp_no IN
(SELECT emp_no FROM salaries WHERE to_date = '9999-01-01')) AS s_hire
INNER JOIN
(SELECT s2.emp_no, s2.salary
 FROM salaries AS s2
 WHERE to_date = '9999-01-01'
) AS s_now
ON s_hire.emp_no = s_now.emp_no
ORDER BY s_now.salary - s_hire.salary
```
